Now we will perform dimentity reduction using Principle component analysis (PCA) in order to obtain princple components.

```python
# since we earlier identified that objid and specobjid are just identifiers for accessing the rows back when they were stored in the
#original databank. Therefore we will not need them for classification as they are not related to the outcome.
solana_df.drop(['objid', 'run', 'rerun', 'camcol', 'field', 'specobjid'], axis=1, inplace=True)
solana_df.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

solana_df_fe = solana_df

# encode class labels to integers
le = LabelEncoder()
y_encoded = le.fit_transform(solana_df_fe['class'])
solana_df_fe['class'] = y_encoded

# Principal Component Analysis
pca = PCA(n_components=3)
ugriz = pca.fit_transform(solana_df_fe[['u', 'g', 'r', 'i', 'z']])

solana_df_fe = pd.concat((solana_df_fe, pd.DataFrame(ugriz)), axis=1)
solana_df_fe.rename({0: 'PCA_1', 1: 'PCA_2', 2: 'PCA_3'}, axis=1, inplace = True)
solana_df_fe.drop(['u', 'g', 'r', 'i', 'z'], axis=1, inplace=True)
solana_df_fe.head()
```

Performing Feature scaling
To scale the values between (0 and 1 ) this will reduce the distortion due to exceptionally high values.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
solana = scaler.fit_transform(solana_df_fe.drop('class', axis=1))
```

Split the dataset into Train test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(solana, solana_df_fe['class'], test_size=0.3)

X_solana_df = pd.DataFrame(solana)
y_solana_df = pd.DataFrame(solana_df_fe['class'])

X_solana_df.to_csv('NN_X_solana.csv', index=False)
y_solana_df.to_csv('NN_y.csv', index=False)
```

```python
print(X_train.shape)
X_train[0:10]
```

```python
print(y_train.shape)
y_train[:10]
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64, input_dim=9, activation='sigmoid'))  # Set input_dim to 9
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(16, activation='sigmoid'))
model.add(Dense(3, activation='softmax'))  # 3 units for 3 classes

# Compile with sparse_categorical_crossentropy if y_train is not one-hot encoded
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Summary of the model
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_split = 0.1, epochs=30, batch_size=32)
```

```python
from sklearn.metrics import accuracy_score

prediction = []

y_predict = model.predict(X_test)

for i in range(len(y_test)):
    prediction.append(np.argmax(y_predict[i]))  # Predictions from model

# Calculate accuracy
acc = accuracy_score(y_test, prediction)  # Directly compare y_test with predictions
print('Accuracy is:', acc * 100, '%')
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay


y_pred = model.predict(X_test)
y_pred = y_pred.argmax(axis=1)

cm = confusion_matrix(y_test, y_pred)

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.show()
```

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop

def build_model(hp):
    model = Sequential()

    # Tune the number of units in the first Dense layer
    hp_units1 = hp.Int('units1', min_value=32, max_value=128, step=32)
    model.add(Dense(units=hp_units1, activation='relu', input_shape=(X_train.shape[1],)))

    # Tune the number of units in the second Dense layer
    hp_units2 = hp.Int('units2', min_value=16, max_value=64, step=16)
    model.add(Dense(units=hp_units2, activation='relu'))

    # Tune the number of units in the third Dense layer
    hp_units3 = hp.Int('units3', min_value=8, max_value=32, step=8)
    model.add(Dense(units=hp_units3, activation='relu'))

    # Output layer
    model.add(Dense(3, activation='softmax'))

    # Tune the learning rate for the optimizer
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(optimizer=RMSprop(learning_rate=hp_learning_rate),
                  loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model

# Instantiate the tuner
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,  # The number of different models to try
    executions_per_trial=2,  # Number of times to train each model for robustness
    directory='hyperparameter_tuning',
    project_name='neural_network_tuning'
)

tuner.search(X_train, y_train, epochs=30, validation_split=0.2, batch_size=32)

best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print("Best hyperparameters:")
print(f"Units in first layer: {best_hps.get('units1')}")
print(f"Units in second layer: {best_hps.get('units2')}")
print(f"Units in third layer: {best_hps.get('units3')}")
print(f"Learning rate: {best_hps.get('learning_rate')}")

# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = best_model.predict(X_test)
y_pred = y_pred.argmax(axis=1)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Galaxies', 'Quasars', 'Stars'],
            yticklabels=['Galaxies', 'Quasars', 'Stars'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()

report = classification_report(y_test, y_pred, target_names=['Galaxies', 'Quasars', 'Stars'])
print("Classification Report:")
print(report)
```
