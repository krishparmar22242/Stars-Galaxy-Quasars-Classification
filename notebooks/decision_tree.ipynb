```python
one_hot_encoded = pd.get_dummies(dataset, columns=['class'])
one_hot_encoded.head()
```

```python
one_hot_encoded = one_hot_encoded.astype(int)
one_hot_encoded.head()
```

```python
one_hot_encoded.describe()
```

```python
numerical_cols = one_hot_encoded.select_dtypes(include=['int']).columns

plt.figure(figsize=(16, 8))

for i, col in enumerate(numerical_cols):
    plt.subplot(2, len(numerical_cols) // 2 + 1, i + 1)
    sns.boxplot(data=one_hot_encoded[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel('')

plt.tight_layout()
plt.show()
```

```python
one_hot_encoded = one_hot_encoded.drop(['objid', 'specobjid'], axis=1)
one_hot_encoded
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Assuming one_hot_encoded dataframe is already loaded
# Separate features and target
X = one_hot_encoded.drop(columns=['class_GALAXY', 'class_QSO', 'class_STAR'])  # Features
y = one_hot_encoded[['class_GALAXY', 'class_QSO', 'class_STAR']].idxmax(axis=1)  # Target column with 0, 1, 2
```

```python
# Initialize the decision tree model
model = DecisionTreeClassifier()

# Apply RFE to select the top 10 features
rfe = RFE(model, n_features_to_select=10)
X_rfe_selected = rfe.fit_transform(X, y)

# Get the selected feature names from RFE
rfe_selected_features = X.columns[rfe.get_support()]
print("RFE selected features:", rfe_selected_features)

X_rfe_selected_df = pd.DataFrame(X_rfe_selected)
y_df = pd.DataFrame(y)

X_rfe_selected_df.to_csv('DT_X_rfe_selected.csv', index=False)
y_df.to_csv('DT_y.csv', index=False)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_rfe_selected, y, test_size=0.2, random_state=42)

# Train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = dt_classifier.predict(X_test)

# Evaluate the model accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy * 100, "%")

# Calculate the prevalence of each class in the test set
unique, counts = np.unique(y_test, return_counts=True)
prevalence = dict(zip(unique, counts / len(y_test)))
print("Prevalence of each class in the test set:", prevalence)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualize the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=['Galaxy', 'Quasar', 'Star'],
            yticklabels=['Galaxy', 'Quasar', 'Star'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

# Classification report
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", class_report)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for tuning
param_grid = {
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 5, 10],
    'criterion': ['gini', 'entropy']
}

# Initialize the Decision Tree model
dt_classifier = DecisionTreeClassifier(random_state=42)

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=dt_classifier, param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters and best estimator
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print("Best Parameters:", best_params)

# Predict using the best model
y_pred_best = best_model.predict(X_test)

# Evaluate the model accuracy
accuracy_best = accuracy_score(y_test, y_pred_best)
print("Tuned Model Accuracy:", accuracy_best * 100, "%")

# Visualize the confusion matrix for the best model
conf_matrix_best = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_best, annot=True, fmt="d", cmap="Blues",
            xticklabels=['Galaxy', 'Quasar', 'Star'],
            yticklabels=['Galaxy', 'Quasar', 'Star'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix for Tuned Model")
plt.show()

# Classification report for the best model
class_report_best = classification_report(y_test, y_pred_best)
print("\nClassification Report for Tuned Model:\n", class_report_best)
```
